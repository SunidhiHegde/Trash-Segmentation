# To Merge multiple category files into a single JSON, follow the code below

```python
paperFile = open('/content/drive/Shareddrives/NN_Project/NN_Final/SortedClasses/paper.json')
medicalFile = open('/content/drive/Shareddrives/NN_Project/NN_Final/SortedClasses/medical_annotated.json')
glassFile = open('/content/drive/Shareddrives/NN_Project/NN_Final/SortedClasses/glass_annotated.json')
eWasteFile = open('/content/drive/Shareddrives/NN_Project/NN_Final/SortedClasses/e-waste.json')
metalFile = open('/content/drive/Shareddrives/NN_Project/NN_Final/SortedClasses/metal_annotated.json')
organicFile = open('/content/drive/Shareddrives/NN_Project/NN_Final/SortedClasses/organic_annotated.json')
```

```python
paper = json.load(paperFile)
medical = json.load(medicalFile)
glass = json.load(glassFile)
eWaste = json.load(eWasteFile)
metal = json.load(metalFile)
organic = json.load(organicFile)
```

```python
classes = [paper, medical, glass, eWaste, metal, organic]
```

# Declare one of the classes as the parent JSON (paper in this case)

```python
masterJson = paper
```

# Category order

1. Paper
2. Medical
3. Glass
4. E-waste
5. Metal
6. Organic

```python
categoryId = 2
for i in range(1, len(classes)):
  class_category = classes[i]
  category = class_category["categories"][0]
  category["id"] = categoryId
  masterJson["categories"].append(category)
  categoryId += 1
```

```python
masterJson["categories"]
```

# Append images for each class with continuous offset of image id

```python
offsetImageId = len(masterJson["images"])
```

```python
offsetImageId
```

```python
for i in range(1, len(classes)):
  offsetImageId = len(masterJson["images"])
  offsetAnnotationId = len(masterJson["annotations"])
  class_image = classes[i]
  categoryId = masterJson["categories"][i]["id"]
  print(categoryId)
  for j in range(0, len(class_image["images"])):
    each_image = class_image["images"][j]
    each_image["id"] = each_image["id"] + offsetImageId
  
  for j in range(0, len(class_image["annotations"])):
    each_annotation = class_image["annotations"][j]
    each_annotation["id"] = each_annotation["id"] + offsetAnnotationId
    each_annotation["image_id"] = each_annotation["image_id"] + offsetImageId
    each_annotation["category_id"] = categoryId

  masterJson["images"].extend(class_image["images"])
  masterJson["annotations"].extend(class_image["annotations"])
```

```python
with open('/content/drive/Shareddrives/NN_Project/NN_Final/SortedClasses/masterJson.json', 'w') as f:
    json.dump(masterJson, f)
```
